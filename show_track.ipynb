```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import uproot
```

```python
# with uproot.open('./build/LXe.root') as file:
with uproot.open('/home/share/zihaoxu/G4_track_simulation/NeutrinoInLXe_E_10.0.root') as file:
    print(file.keys())
    df = pd.DataFrame({
        'event_id': np.array(file['LXe']['Event']),
        'x': np.array(file['LXe']['X']),
        'y': np.array(file['LXe']['Y']),
        'z': np.array(file['LXe']['Z']),
        'e': -np.array(file['LXe']['Ed']),
    })
```

```python
mask = df['event_id'] == 0
print(df[mask]['e'].sum())

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
ax.scatter(df[mask]['x'], df[mask]['y'], df[mask]['z'])
plt.show()
```
